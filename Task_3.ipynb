1. Maximum Product of Two Elements in an Array

```python
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        nums.sort()
        return (nums[-1] - 1) * (nums[-2] - 1)
```

2. Count Number of Teams

```python
class Solution:
    def numTeams(self, rating: List[int]) -> int:
        n = len(rating)
        teams = 0
        for j in range(n):
            left_smaller = left_greater = 0
            right_smaller = right_greater = 0
            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                elif rating[i] > rating[j]:
                    left_greater += 1
            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_greater += 1
                elif rating[k] < rating[j]:
                    right_smaller += 1

            teams += left_smaller * right_greater
            teams += left_greater * right_smaller
        return teams
```

3. No of Students Doing Homework at a given Time

```python
class Solution:
    def busyStudent(self, startTime: List[int], endTime: List[int], queryTime: int) -> int:
        count = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1
        return count
```

4. No of Steps toRreduce a Number to Zero

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        steps = 0
        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1
        return steps
```

5. Counting Bits

```python
class Solution:
    def countBits(self, n: int) -> List[int]:
        a = [0] * (n + 1)
        for i in range(1, n + 1):
            a[i] = a[i // 2] + (i % 2)
        return a
```
